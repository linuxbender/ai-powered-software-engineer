### Exercise: Singleton Configuration Manager  

**Goal:** Implement a Singleton `ConfigManager` class to manage application-wide settings.  

**Steps:**  
1. Create a class `ConfigManager`.  
2. Ensure that only **one instance** of `ConfigManager` can exist.  
3. Add a dictionary `settings` to store key-value pairs.  
4. Implement methods:  
   - `set(key, value)` → stores a setting.  
   - `get(key)` → retrieves a setting.  

**Behavior to Verify:**  
```python
config1 = ConfigManager()
config2 = ConfigManager()

config1.set("theme", "dark")
print(config2.get("theme"))   # "dark" -  Both share same settings

print(config1 is config2)     # True - Only one instance

````python
class ConfigManager:
    _instance = None

    def __new__(cls, *args, **kwargs):
        if cls._instance is None:
            cls._instance = super(ConfigManager, cls).__new__(cls)
        return cls._instance

    def __init__(self):
        if not hasattr(self, "settings"):
            self.settings = {}

    def set(self, key, value):
        self.settings[key] = value

    def get(self, key):
        if key not in self.settings:
            raise KeyError(f"Setting '{key}' not found.")
        return self.settings[key]
````

````python
import unittest

class TestSingletonConfigManager(unittest.TestCase):

    def test_single_instance(self):
        config1 = ConfigManager()
        config2 = ConfigManager()
        self.assertIs(config1, config2)

    def test_set_and_get(self):
        config = ConfigManager()
        config.set("theme", "dark")
        self.assertEqual(config.get("theme"), "dark")

    def test_shared_state(self):
        config1 = ConfigManager()
        config2 = ConfigManager()
        config1.set("language", "English")
        self.assertEqual(config2.get("language"), "English")

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
````
