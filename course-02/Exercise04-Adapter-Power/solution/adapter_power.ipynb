### Exercise: Power Adapter Pattern  

**Goal:** Implement the Adapter Pattern to make a `USPlug` compatible with a `EuropeanSocket`.  

**Steps:**  
1. Understand the incompatibility:  
   - `EuropeanSocket` provides a method `plug_in()` returning `"220V power from European socket"`.  
   - `USPlug` expects a method `connect()` returning `"110V power device connected"`.  

2. Create a class `USPlugAdapter` that:  
   - Takes a `EuropeanSocket` object in its constructor.  
   - Implements a method `connect()` that adapts the socket’s `plug_in()` output.  

3. Inside `connect()`, call `european_socket.plug_in()` and return a string showing the conversion, for example:  "Adapter converts -> 220V power from European socket to 110V"


4. Test the adapter by:  
- Instantiating a `EuropeanSocket`.  
- Passing it into `USPlugAdapter`.  
- Calling `connect()` on the adapter to confirm it behaves like a `USPlug`. 

```python
# Incompatible interface
class EuropeanSocket:
    def plug_in(self):
        return "220V power from European socket"

# Desired interface
class USPlug:
    def connect(self):
        return "110V power device connected"

class USPlugAdapter(USPlug):
    def __init__(self, european_socket: EuropeanSocket):
        self.european_socket = european_socket

    def connect(self):
        # Adapts the European socket to behave like a US plug
        return f"Adapter converts -> {self.european_socket.plug_in()} to 110V"
        

# Example usage
if __name__ == "__main__":
    euro_socket = EuropeanSocket()
    adapter = USPlugAdapter(euro_socket)

    print(adapter.connect())  
```

```python
import unittest

class TestAdapterPattern(unittest.TestCase):
    def test_us_plug_direct(self):
        plug = USPlug()
        self.assertEqual(plug.connect(), "110V power device connected")

    def test_adapter_output(self):
        euro_socket = EuropeanSocket()
        adapter = USPlugAdapter(euro_socket)
        self.assertEqual(
            adapter.connect(),
            "Adapter converts -> 220V power from European socket to 110V"
        )

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
