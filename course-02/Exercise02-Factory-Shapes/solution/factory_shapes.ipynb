### Exercise: Factory Pattern in Python  

**Goal:** Implement the Factory Method Pattern to create shapes (`Circle`, `Square`, `Triangle`) with a `draw()` method.  

**Steps:**  
1. Create an abstract class `Shape` with an abstract method `draw()`.  
2. Implement the following classes:  
   - `Circle.draw()` returns string `"Drawing a Circle"`  
   - `Square.draw()` returns string `"Drawing a Square"`  
   - `Triangle.draw()` returns string `"Drawing a Triangle"`  
3. Complete `ShapeFactory.get_shape(shape_type)` so that:  
   - `"circle"` input returns  string `"Drawing a Circle"`  
   - `"square"` input returns string `"Drawing a Square"`   
   - `"triangle"` input returns string `"Drawing a Square"` 
   - Any other input → raise `ValueError("Invalid shape type")`  

**Test:**  
Run the provided test cell. All tests should pass.


```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def draw(self):
        pass


class Circle(Shape):
    def draw(self):
        return "Drawing a Circle"


class Square(Shape):
    def draw(self):
        return "Drawing a Square"


class Triangle(Shape):
    def draw(self):
        return "Drawing a Triangle"


class ShapeFactory:
    def get_shape(self, shape_type):
        if shape_type is None:
            return None
        shape_type = shape_type.lower()
        if shape_type == "circle":
            return Circle()
        elif shape_type == "square":
            return Square()
        elif shape_type == "triangle":
            return Triangle()
        else:
            raise ValueError(f"Unknown shape type: {shape_type}")


# Example usage:
if __name__ == "__main__":
    factory = ShapeFactory()

    shapes = ["circle", "square", "triangle", "pentagon"]

    for s in shapes:
        shape = factory.get_shape(s)
        if shape:
            print(shape.draw())
        else:
            print(f"No shape found for: {s}")
```

```python
import unittest

class TestShapeFactory(unittest.TestCase):
    def setUp(self):
        self.factory = ShapeFactory()

    def test_circle(self):
        shape = self.factory.get_shape("circle")
        self.assertEqual(shape.draw(), "Drawing a Circle")

    def test_square(self):
        shape = self.factory.get_shape("square")
        self.assertEqual(shape.draw(), "Drawing a Square")

    def test_triangle(self):
        shape = self.factory.get_shape("triangle")
        self.assertEqual(shape.draw(), "Drawing a Triangle")

    def test_invalid_shape(self):
        with self.assertRaises(ValueError):
            self.factory.get_shape("hexagon")

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
