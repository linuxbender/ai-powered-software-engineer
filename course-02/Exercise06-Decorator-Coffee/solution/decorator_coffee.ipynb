### Exercise: Decorator Pattern – Coffee Shop

**Goal:** Implement the **Decorator Pattern** to dynamically add features (like condiments) to a `Coffee` object.  
The Decorator Pattern allows **adding responsibilities to objects at runtime** without modifying the original class.

---

**Steps:**  

1. Create an abstract class `Coffee` with two abstract methods:  
   - `get_cost()` → returns the cost of the coffee  
   - `get_ingredients()` → returns a string listing the ingredients  

2. Implement a `SimpleCoffee` class (Concrete Component):  
   - `get_cost()` returns a base price (e.g., 5)  
   - `get_ingredients()` returns `"Coffee"`  

3. Create an abstract decorator class `CoffeeDecorator`:  
   - Inherits from `Coffee`  
   - Stores a reference to a `Coffee` object (`self.component`)  

4. Implement concrete decorators (e.g., `MilkDecorator`, `SugarDecorator`, `WhipDecorator`):  
   - Each overrides `get_cost()` to add its price  
   - Each overrides `get_ingredients()` to append its ingredient  

5. Test your implementation by wrapping `SimpleCoffee` in multiple decorators and verifying:  
   - The total cost  
   - The full ingredient list  

```python
from abc import ABC, abstractmethod

# Step 1: Component
class Coffee(ABC):
    @abstractmethod
    def get_cost(self):
        pass

    @abstractmethod
    def get_ingredients(self):
        pass

# Step 2: Concrete Component
class SimpleCoffee(Coffee):
    def get_cost(self):
        return 5  # base price

    def get_ingredients(self):
        return "Coffee"

# Step 3: Decorator
class CoffeeDecorator(Coffee):
    def __init__(self, coffee):
        self.component = coffee  # store reference to decorated coffee

    @abstractmethod
    def get_cost(self):
        pass

    @abstractmethod
    def get_ingredients(self):
        pass

# Step 4: Concrete Decorators
class MilkDecorator(CoffeeDecorator):
    def get_cost(self):
        return self.component.get_cost() + 2  # Milk costs 2

    def get_ingredients(self):
        return self.component.get_ingredients() + ", Milk"


class SugarDecorator(CoffeeDecorator):
    def get_cost(self):
        return self.component.get_cost() + 1  # Sugar costs 1

    def get_ingredients(self):
        return self.component.get_ingredients() + ", Sugar"


class WhipDecorator(CoffeeDecorator):
    def get_cost(self):
        return self.component.get_cost() + 3  # Whip costs 3

    def get_ingredients(self):
        return self.component.get_ingredients() + ", Whip"

# Example Usage
if __name__ == "__main__":
    coffee = SimpleCoffee()
    coffee = MilkDecorator(coffee)
    coffee = SugarDecorator(coffee)
    coffee = WhipDecorator(coffee)

    print(coffee.get_cost())        # Output: 5 + 2 + 1 + 3 = 11
    print(coffee.get_ingredients()) # Output: Coffee, Milk, Sugar, Whip
```

```python
import unittest

class TestDecoratorPattern(unittest.TestCase):
    def test_simple_coffee(self):
        coffee = SimpleCoffee()
        self.assertEqual(coffee.get_cost(), 5)
        self.assertEqual(coffee.get_ingredients(), "Coffee")

    def test_milk_and_sugar(self):
        coffee = MilkDecorator(SugarDecorator(SimpleCoffee()))
        self.assertEqual(coffee.get_ingredients(), "Coffee, Sugar, Milk")
        self.assertEqual(coffee.get_cost(), 5 + 2 + 1)  # assuming Sugar=1, Milk=2

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
