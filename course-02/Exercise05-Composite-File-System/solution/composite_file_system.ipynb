### Exercise: Composite File System

**Goal:** Implement the **Composite Design Pattern** to represent a file system where both **files** and **directories** can be treated uniformly.  

**Steps:**  
1. Create an abstract class `FileSystemComponent` with two abstract methods:  
   - `show_details(indent=0)` → prints the component’s details with indentation.  
   - `get_size()` → returns the size of the component.  

2. Implement the **Leaf class** `File`:  
   - Stores a `name` and `size` (in KB).  
   - `show_details(indent)` prints the file name and size with indentation.  
   - `get_size()` returns the file’s size.  

3. Implement the **Composite class** `Directory`:  
   - Stores a `name` and a list of children (`File` or `Directory`).  
   - `add(component)` adds a file or directory to its children.  
   - `show_details(indent)` prints the directory name and recursively prints all children with extra indentation.  
   - `get_size()` returns the total size of all children.  

4. In the main section, build a small file system:  
   - A root directory containing two files and a subdirectory.  
   - The subdirectory contains another file.  

```python
from abc import ABC, abstractmethod

# Step 1: Abstract Component
class FileSystemComponent(ABC):
    @abstractmethod
    def show_details(self, indent=0):
        pass

    @abstractmethod
    def get_size(self):
        pass


# Step 2: Leaf (File)
class File(FileSystemComponent):
    def __init__(self, name, size):
        self.name = name
        self.size = size

    def show_details(self, indent=0):
        # Print the file name with indentation
        print("  " * indent + f"File: {self.name} ({self.size} KB)")

    def get_size(self):
        # Return the size of the file
        return self.size


# Step 3: Composite (Directory)
class Directory(FileSystemComponent):
    def __init__(self, name):
        self.name = name
        self.children = []

    def add(self, component):
        # Add file or directory to children
        self.children.append(component)

    def show_details(self, indent=0):
        # Print directory name
        print("  " * indent + f"Directory: {self.name}")
        # Recursively print children with extra indentation
        for child in self.children:
            child.show_details(indent + 1)

    def get_size(self):
        # Total size = sum of all children sizes
        return sum(child.get_size() for child in self.children)


if __name__ == "__main__":
    root = Directory("root")
    file1 = File("file1.txt", 100)
    file2 = File("file2.txt", 200)

    subdir = Directory("subdir")
    subdir.add(File("file3.txt", 300))

    root.add(file1)
    root.add(file2)
    root.add(subdir)

    root.show_details()
    print("Total size:", root.get_size())
```

```python
import unittest

class TestCompositePattern(unittest.TestCase):
    def test_file_size(self):
        file = File("test.txt", 150)
        self.assertEqual(file.get_size(), 150)

    def test_directory_size(self):
        root = Directory("root")
        root.add(File("a.txt", 100))
        root.add(File("b.txt", 200))
        self.assertEqual(root.get_size(), 300)

    def test_nested_directory(self):
        root = Directory("root")
        subdir = Directory("subdir")
        subdir.add(File("c.txt", 300))
        root.add(subdir)
        self.assertEqual(root.get_size(), 300)

if __name__ == "__main__":
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
